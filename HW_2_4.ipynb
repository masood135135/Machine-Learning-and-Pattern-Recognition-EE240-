# Loading Data and separating 4 and 9 numbers

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import fetch_mldata
#downloding mnist data
mnist = fetch_mldata('MNIST original')

X = mnist.data
y = mnist.target

X4=X[y==4];  X9=X[y==9];
y4=y[y==4]; y9=y[y==9];


X4_train=X4[0:4000,:];  X9_train=X9[0:4000,:]; 
X4_test=X4[4000:,:]; X9_test=X9[4000:,:]

y4_train=y4[0:4000];  y9_train=y9[0:4000]; 
y4_test=y4[4000:]; y9_test=y9[4000:]

X_train=np.vstack((X4_train,X9_train));
y_train=np.concatenate((y4_train,y9_train));

X_test=np.vstack((X4_test,X9_test));
y_test=np.concatenate((y4_test,y9_test));
```

# Finding Optimum C for kernel='poly' and degree=1

```python

#degree 1

clf= svm.SVC(C=0.001,kernel='poly',degree=1)
clf.fit(X_train,y_train)
S_V=clf.support_vectors_

Pe = 1 - clf.score(X_test,y_test)

print('Support Vector ={}   Error={} for C=0.001'.format(len(S_V),Pe))


clf= svm.SVC(C=0.01,kernel='poly',degree=1)
clf.fit(X_train,y_train)
S_V=clf.support_vectors_
Pe = 1 - clf.score(X_test,y_test)
print('Support Vector ={}   Error={} for C=0.01'.format(len(S_V),Pe))


clf= svm.SVC(C=0.1,kernel='poly',degree=1)
clf.fit(X_train,y_train)
S_V=clf.support_vectors_
Pe = 1 - clf.score(X_test,y_test)
print('Support Vector ={}   Error={} for C=0.1'.format(len(S_V),Pe))


clf= svm.SVC(C=1,kernel='poly',degree=1)
clf.fit(X_train,y_train)
S_V=clf.support_vectors_
Pe = 1 - clf.score(X_test,y_test)
print('Support Vector ={}   Error={} for C=1'.format(len(S_V),Pe))


clf= svm.SVC(C=10,kernel='poly',degree=1)
clf.fit(X_train,y_train)
S_V=clf.support_vectors_
Pe = 1 - clf.score(X_test,y_test)
print('Support Vector ={}   Error={} for C=10'.format(len(S_V),Pe))




```
