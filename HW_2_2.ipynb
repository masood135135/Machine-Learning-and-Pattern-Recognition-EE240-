```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

# part a

```python
x=np.random.rand(100,2)
x1=np.ones(100).reshape(100,1)
w=np.ones(2).reshape(2,1)
y=np.sign(x.dot(w)-1)
Data=np.hstack((y,x1,x))
Data1_train=Data[0:80,:]
Data1_test=Data[81:100,:]
Data2_train=Data[0:30,:]
Data2_test=Data[31:100,:]
```

# part b

```python
w=np.array([[0,0,0]]).reshape(3,1)
def h(w,X):
    return np.sign(X.dot(w))

for i in range (0,80):
        w = w + 0.5*(Data1_train[i,0]-h(w,Data1_train[i,1:4]))*Data1_train[i,1:4].reshape(3,1);
print(w)
```

# part d

```python
y_cal=np.sign((Data1_train[:,1:4].dot(w)))
y_cor=np.sign(Data1_train[:,2:4].dot(np.array([1, 1]))-1)
tp=y_cal==np.ones(80).reshape(80,1)*1
tp_b=(y_cal==y_cor)*1
tp=tp_b*tp

fp=y_cal==np.ones(80).reshape(80,1)*1
fp_b=(y_cal!=y_cor)*1
fp=fp_b*fp

fn=y_cal==np.ones(80).reshape(80,1)*-1
fn_b=(y_cal!=y_cor)*1
fn=fn_b*fn

Precision=np.sum(tp)/(np.sum(tp)+np.sum(fp))

Recall=np.sum(tp)/(np.sum(tp)+np.sum(fn))

f1_score=2*(Precision*Recall)/(Precision+Recall)

print('Precision={}\nRecall={}\nf1_score={}'.format(Precision,Recall,f1_score))

```
